```python
from bs4 import BeautifulSoup
import pandas as pd

"""
url = "https://www.linkedin.com/jobs/search?keywords=Data%2BScientist&location=India&trk=public_jobs_jobs-search-bar_search-submit&currentJobId=3983495628&position=1&pageNum=0&original_referer=https%3A%2F%2Fwww.linkedin.com%2Fjobs%2Fsearch%3Fkeywords%3DData%252BScientist%26location%3DIndia%26trk%3Dpublic_jobs_jobs-search-bar_search-submit%26currentJobId%3D3970018098%26position%3D3%26pageNum%3D2"

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")
"""

with open('linkedin_jobs_page.html', 'r', encoding='utf-8') as f:
    soup = BeautifulSoup(f, 'html.parser')

# Lists to store job data
job_titles = []
companies = []
locations = []
post_times = []
links = []

# Attempting to find all job listings by targeting the appropriate divs
job_cards = soup.find_all("div", class_="base-card")

# Extract information from each job card
for card in job_cards:
    # Extract job title
    title_tag = card.find("h3", class_="base-search-card__title")
    title = title_tag.text.strip() if title_tag else "N/A"
    
    # Extract company name
    company_tag = card.find("h4", class_="base-search-card__subtitle")
    company = company_tag.text.strip() if company_tag else "N/A"
    
    # Extract job location
    location_tag = card.find("span", class_="job-search-card__location")
    location = location_tag.text.strip() if location_tag else "N/A"
    
    # Extract time of posting
    post_time_tag = card.find("time")
    post_time = post_time_tag.text.strip() if post_time_tag else "N/A"
    
    # Extract job link
    link_tag = card.find("a", class_="base-card__full-link")
    link = link_tag['href'].strip() if link_tag else "N/A"
    
    # Append data to lists
    job_titles.append(title)
    companies.append(company)
    locations.append(location)
    post_times.append(post_time)
    links.append(link)

# Create a DataFrame
jobs_df = pd.DataFrame({
    "Job Title": job_titles,
    "Company": companies,
    "Location": locations,
    "Time of Posting": post_times,
    "Job Link": links
})

# Save to CSV
jobs_df.to_csv("linkedin_jobs.csv", index=False)
print("Job listings saved to linkedin_jobs.csv")

```

```python
from bs4 import BeautifulSoup
import re

# Load the HTML file
with open("job_description_page_test.html", "r", encoding="utf-8") as file:  # Update with your file path
    soup = BeautifulSoup(file, "html.parser")

# Find the section with the class 'show-more-less-html'
section = soup.find("div", class_="show-more-less-html__markup show-more-less-html__markup--clamp-after-5 relative overflow-hidden")

# Function to extract specific job-related information
def extract_job_details():
    details = {}
    
    # Look for specific labels
    labels_to_find = ["Seniority level", "Employment type", "Job function", "Industries"]
    
    for label in labels_to_find:
        # Find the label in the section
        label_element = soup.find("h3", string=re.compile(label))
        if label_element:
            # Find the next sibling containing the relevant data
            value_element = label_element.find_next("span")
            details[label] = value_element.get_text(strip=True) if value_element else "Not Found"

    return details

if section:
    # Extract heading-tagged content
    
    job_details = section.get_text()
    # Print the extracted heading-tagged information
    print("Extracted job Description:\n" + "-"*40)
    print(job_details)

    # Extract job-related details
    job_details = extract_job_details()
    print("\nExtracted Job Details:\n" + "-"*40)
    for label, value in job_details.items():
        print(f"{label}: {value}")
else:
    print("The specified section 'show-more-less-html' was not found.")
```

```python
# Initialize the dictionary to store the extracted information
extracted_info = {}

# Extract paragraphs and list items under each heading
current_heading = None

for tag in section.find_all(["strong", "li", "p", "br"]):
    # Handle headings
    if tag.name == "strong":
        heading = tag.get_text(strip=True).replace("\n", "")
        current_heading = heading
        extracted_info[current_heading] = {"description": "", "list": []}
    # Handle list items
    elif tag.name == "li" and current_heading:
        extracted_info[current_heading]["list"].append(tag.get_text(strip=True))
    # Handle general text under the current heading
    elif tag.name in ["p", "br"] and current_heading:
        extracted_info[current_heading]["description"] += tag.get_text(strip=True) + " "

# Clean up extra spaces and empty sections
for heading, content in extracted_info.items():
    extracted_info[heading]["description"] = content["description"].strip()
    if not extracted_info[heading]["list"]:
        del extracted_info[heading]["list"]

# Print the extracted information
for heading, content in extracted_info.items():
    print(f"Heading: {heading}")
    print(f"Description: {content['description']}")
    if "list" in content:
        print("List Items:")
        for item in content["list"]:
            print(f"- {item}")
    print("-" * 40)

# The final structured information is stored in `extracted_info` dictionary
```
